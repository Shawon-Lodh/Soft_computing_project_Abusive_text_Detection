#LSTM

```
import numpy as np
import matplotlib.pyplot as plt
import math

N = 6
ind = np.arange(N)  # the x locations for the groups
width = 0.30       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)

yvals = [93.57,94.57,94.25,94.61,95.21,95.08]
rects1 = ax.bar(ind, yvals, width, color='#016ec2')
zvals = [93.37,94.75,94.95,94.72,95.16,95.11]
rects2 = ax.bar(ind+width, zvals, width, color='#fe8400')

low = min(yvals)-1
high = max(yvals)-1
plt.ylim([math.ceil(low-0*(high-low)),  math.ceil(high+0.5*(high-low))])


ax.set_ylabel('Accuracy in %')
ax.set_xticks(ind+(width/2))
ax.set_xticklabels( ('Normal_trainable', 'glove-twitter-25', 'glove-twitter-50', 'glove-twitter-100', 'glove-twitter-200', 'fasttext-wiki-news-\nsubwords-300') )
ax.set_title('Accuracy of LSTM by different features') 
ax.legend( (rects1[0], rects2[0]), ('Single Direction', 'Bi-Direction') )

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}%'.format(height),
                    xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
plt.rcParams['figure.figsize'] = (15,8)
plt.show()
```
